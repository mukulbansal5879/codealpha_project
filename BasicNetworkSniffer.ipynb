```python
from scapy.all import *
```

```python
def create_custom_packet():
    src_ip = input("Enter Source IP Address: ")
    dst_ip = input("Enter Destination IP Address: ")
    protocol = input("Enter Protocol (TCP/UDP/ICMP): ").upper()
    payload = input("Enter Payload (optional): ")
```

```python
def analyze_packets(packet_list):
    for i, pkt in enumerate(packet_list):
        print(f"\n📦 Packet {i+1}:")
        if IP in pkt:
            print(f"  - Source IP       : {pkt[IP].src}")
            print(f"  - Destination IP  : {pkt[IP].dst}")
            print(f"  - Protocol        : {pkt.proto}")
        if TCP in pkt:
            print("  - TCP Detected    : Port", pkt[TCP].dport)
        elif UDP in pkt:
            print("  - UDP Detected    : Port", pkt[UDP].dport)
        elif ICMP in pkt:
            print("  - ICMP Detected")

        if Raw in pkt:
            print(f"  - Payload         : {pkt[Raw].load.decode(errors='ignore')}")
        else:
            print("  - No Payload")
```

```python
packet = create_custom_packet()
if packet:
    analyze_packets([packet])
```
